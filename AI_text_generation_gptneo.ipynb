```python
from transformers import pipeline
```

**Setup Generator**

```python
generator = pipeline('text-generation', model='EleutherAI/gpt-neo-2.7B')
```

**Generate Text using Prompt**

```python
prompt="Bitcoin is going to"
```

```python
text=generator(prompt, max_length=50, do_sample=True, temperature=0.9)
#temperature :val used to model next set of probabilities
#prompt:text used start generation
#max_length: len of text to be generated
#do_sample:whether or not to use sampling
```

**Output text**

```python
 print(text[0]['generated_text'])
```

```python
with open('gpttext.txt', 'w') as f:
  f.writelines(text[0]['generated_text'])
```
